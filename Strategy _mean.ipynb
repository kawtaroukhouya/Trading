```python
import pandas as pd
import numpy as np
import datetime
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
```

```python
df = pd.read_excel('3Y_Data_SP500.xlsx')
df.head()
```

```python
df.dtypes
```

```python
df["Open"] = [float(str(i).replace(",", "")) for i in df["Open"]]
```

```python
df["High"] = [float(str(i).replace(",", "")) for i in df["High"]]
```

```python
df["Low"] = [float(str(i).replace(",", "")) for i in df["Low"]]
```

```python
df["Close"] = [float(str(i).replace(",", "")) for i in df["Close"]]
```

```python
df['Date'] = df["Date"].astype('datetime64[ns]')
```

```python
df['diffHigh']=df['High']-df['Open']
```

```python
df['difflow']=df['Open']-df['Low']
```

```python
df['diffClose']=df['Close']-df['Open']
```

```python
df.head()
```

```python
df['Average'] = (df['diffHigh'] + df['difflow']+df['diffClose'])/3
df.head()
```

```python
df['Average'].mean()
```

```python
df['diffHigh'].mean()
```

```python
df['Target_price1']=df['Open']+13.43
```

```python
df['Target_price2']=df['Open']-13.43
```

```python
df=df.drop(columns=['Target_price','Gain_lost','Target_price2'])
```

```python
df=df.drop(columns=['Gain_lost1','Target_price1','Gain_lost2'])
```

```python
df.head()
```

```python
df['Gain_lost1'] = np.select([df['Target_price1']<df['High'],df['Target_price1']>df['High']], ['Gain','Lost'])
df.head()
```

```python
df['Gain_lost2'] = np.select([df['Target_price2']<df['Low'],df['Target_price2']>df['Low']], ['Lost','Gain'])
df.head()
```

```python
df["Gain_lost1"].value_counts()
```

```python
df["Gain_lost2"].value_counts()
```

Accuracy= 49%
BASED ON THIS WORK THE ACCURACY OF OUR ALGORITHM IS 49% WHICH IS CALCULATED BY DIVIDING THE NUMBER OF Gain BY THE TOTAL NUMBER OF OBSERVATIONS 503 THEN MULTIPLY BY 100

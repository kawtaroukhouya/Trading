```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load in the S&P 500 data
data = pd.read_csv("SP5002.csv")
```

```python
data["Close"] = [float(str(i).replace(",", "")) for i in data["Close"]]
```

```python
# Calculate the 50-day moving average
data["50MA"] = data["Close"].rolling(5).mean()
```

```python
# Calculate the 200-day moving average
data["200MA"] = data["Close"].rolling(20).mean()
```

```python
# Plot the 50-day and 200-day moving averages
plt.plot(data["Date"], data["50MA"], label="50MA")
plt.plot(data["Date"], data["200MA"], label="200MA")
plt.legend(loc="upper left")
plt.show()
```

```python
# Identify when the 50MA crosses above the 200MA (a buy signal)
buy_signals = data[data["50MA"] > data["200MA"]]
```

```python
# Identify when the 50MA crosses below the 200MA (a sell signal)
sell_signals = data[data["50MA"] < data["200MA"]]
```

```python
data.tail()
```

```python
data.loc[:,'Strategy'] = 'buy'
data.loc[data['50MA']<data['200MA'],'Strategy'] = 'sell'
data.head()
```

```python
data.isnull().sum()
```

```python
data = data.dropna()
```

```python
data.sort_values(by='Date', inplace = True)
```

```python
df = pd.read_csv("SP5002.csv")
```

```python
df["Close"] = [float(str(i).replace(",", "")) for i in df["Close"]]
df["Open"] = [float(str(i).replace(",", "")) for i in df["Open"]]
df["Low"] = [float(str(i).replace(",", "")) for i in df["Low"]]
df["High"] = [float(str(i).replace(",", "")) for i in df["High"]]
```

```python
df.loc[:,'Actual'] = 'buy'
df.loc[df['Close']<df['Open'],'Actual'] = 'sell'
df.head()
```

```python
df.sort_values(by='Date', inplace = True)
```

```python
df.head()
```

```python
df.drop(columns = ['Open','High','Low','Close'], inplace=True)
```
